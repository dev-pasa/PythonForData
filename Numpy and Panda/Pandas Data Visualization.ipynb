# Examples of Pandas Built-in Plot


* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

```python
#Area Plot
df2.plot(alpha=3)
```

```python
#Bar Plot
df2.plot.bar()
```

```python
#Bar Plot
df2.plot.bar(stacked=True)
```

```python
#Histograms
df1['A'].plot.hist(bins=50)
```

```python
#Line Plots
df1.plot.line(x=df1.index, y='B', figsize=(12,3), lw=1)
```

```python
#Scatter Plots
df1.plot.scatter(x='A', y='B')
```

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
